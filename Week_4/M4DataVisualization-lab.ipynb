```python
# print all the tables names in the database
QUERY = """
SELECT name as Table_Name FROM
sqlite_master WHERE
type = 'table'
"""
# the read_sql_query runs the sql query and returns the data as a dataframe
pd.read_sql_query(QUERY,conn)
```

```python
QUERY = """
SELECT Age,COUNT(*) as count
FROM master
group by age
order by age
"""
pd.read_sql_query(QUERY,conn)
```

# Hands-on Lab


## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# your code goes here
QUERY = ''' SELECT ConvertedComp FROM master '''

df = pd.read_sql_query(QUERY, conn)
plt.hist(df)
plt.show()
```

### Box Plots


Plot a box plot of `Age.`


```python
# your code goes here
QUERY = ''' SELECT Age FROM master '''

df = pd.read_sql_query(QUERY, conn)
sns.boxplot(df)
plt.ylabel('Age')
plt.title('Age Distribution')
plt.show()
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
# your code goes here
QUERY = ''' SELECT Age, WorkWeekHrs FROM master '''

df = pd.read_sql_query(QUERY, conn)
sns.scatterplot(x='Age', y='WorkWeekHrs', data=df)
plt.show()
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
# your code goes here

QUERY = ''' SELECT Age, WorkWeekHrs, CodeRevHrs FROM master '''

df = pd.read_sql_query(QUERY, conn)
plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x='CodeRevHrs', y='WorkWeekHrs', size='Age', alpha=0.6, sizes=(1,1000))
plt.show()
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
QUERY = '''SELECT *, COUNT(*) AS count FROM DatabaseDesireNextYear GROUP BY DatabaseDesireNextYear ORDER BY count DESC LIMIT 5'''
df = pd.read_sql_query(QUERY, conn)
data=df['count']
labels=df['DatabaseDesireNextYear']
plt.figure(figsize=(8,8))
plt.pie(data, labels = labels, autopct='%1.1f%%', explode = (0.1, 0, 0, 0, 0), shadow = True)
plt.show()
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
# your code goes here
QUERY = '''SELECT Age,WorkWeekHrs, CodeRevHrs FROM master'''
# Load the SQLite database into a pandas DataFrame
df = pd.read_sql(QUERY, conn)

# Filter data for the age group 30 to 35
filtered_data = df[(df['Age'] >= 30) & (df['Age'] <= 35)]

# Group the filtered data by 'Age' and calculate median 'WorkWeekHrs' and 'CodeRevHrs'
grouped_data = filtered_data.groupby('Age').median()[['WorkWeekHrs', 'CodeRevHrs']]

# Plot the stacked bar chart
grouped_data.plot(kind='bar', stacked=True)

# Add labels and title
plt.title('Median WorkWeekHrs and CodeRevHrs for Age Group 30 to 35')
plt.xlabel('Age')
plt.ylabel('Median Hours')

# Show the plot
plt.show()
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
# your code goes here
QUERY = ''' SELECT Age, ConvertedComp FROM master '''

df = pd.read_sql_query(QUERY, conn)
filtered_data = df[(df['Age'] >= 45) & (df['Age'] <= 60)]

# Group the filtered data by 'Age' and calculate median 'ConvertedComp'
median_comp_by_age = filtered_data.groupby('Age')['ConvertedComp'].median()

# Plot the median 'ConvertedComp' for each age
plt.figure(figsize=(10, 6))
median_comp_by_age.plot(kind='line', marker='o')

# Add labels and title
plt.title('Median ConvertedComp for Ages 45 to 60')
plt.xlabel('Age')
plt.ylabel('Median ConvertedComp')

# Show the plot
plt.grid(True)
```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python
# your code goes here
QUERY = ''' SELECT * FROM master '''

df = pd.read_sql_query(QUERY, conn)

# Count occurrences of each category in 'MainBranch'
main_branch_counts = df['MainBranch'].value_counts()

# Plot the horizontal bar chart
plt.figure(figsize=(10, 6))
main_branch_counts.plot(kind='barh')

# Add labels and title
plt.title('Main Branch Distribution')
plt.xlabel('Count')
plt.ylabel('Main Branch')

# Show the plot
plt.show()
```
